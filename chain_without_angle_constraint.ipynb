```python
import os
import jax
import jax.numpy as jnp
import matplotlib.pyplot as plt
from jax import grad, jit, vmap
from IPython import display
```

```python
# global parameters
g=10.0  # pulling force on first bead
m=1.0  # mass
k=1e2  # spring constant
d0=1e-2  # bond length
```

```python
# potential
def V(X):   
    # hookean spring chain potential
    d = jnp.sqrt(jnp.sum(jnp.square(X[1:] - X[:-1]), axis=1))
    E1 = 0.5 * k * jnp.sum(jnp.square(d - d0))

    return E1

# negative force function
nF = grad(V)

# acceleration
def A(X):
    # acceleration from potential
    A0 = -nF(X)/m
    
    # add pulling force for first bead
    A0 = jax.ops.index_add(A0, (0,1), -g)
    
    return A0
```

```python
# parameters
N = 1000
dt = 1e-3
num_steps = 2000

# initial positions and velocities
Xi = jnp.zeros((N, 2))
Vi = jnp.zeros((N, 2))

# lay out beads
offset = jnp.linspace(0.0, -N*d0, N)
for i in range(N):
    Xi = jax.ops.index_add(Xi, (i,0), offset[i])

# set first bead coordinates with offset
Xi = jax.ops.index_add(Xi, (0,0), d0)
Xi = jax.ops.index_add(Xi, (0,1), -d0)

for i in range(num_steps):
    # leapfrog integration
    Ai = A(Xi)
    Xip1 = Xi + Vi * dt + 0.5 * Ai * dt * dt
    Aip1 = A(Xip1)
    Vip1 = Vi + 0.5 * (Ai + Aip1) * dt

    # update positions and velocities
    Xi = Xip1
    Vi = Vip1
    
    # plot
    if (i % 10) == 0:
        plt.gca().cla() 
        plt.plot(Xi[:,0], Xi[:,1], 'o-')
        plt.plot([-1.0,0.2], [0.0,0.0], 'k--')
        plt.xlim(-1.0,0.2)
        plt.ylim(-1.0,0.2)
        plt.xlabel('x')
        plt.ylabel('y')
        plt.title("step = {}".format(i))
        display.clear_output(wait=True)
        display.display(plt.gcf()) 
```

```python
plt.figure()
plt.plot(Xi[:,0], Xi[:,1], 'o-')
plt.plot([-1.0,0.2], [0.0,0.0], 'k--')
plt.xlim(-1.0,0.2)
plt.ylim(-1.0,0.2)
plt.xlabel('x')
plt.ylabel('y')
plt.title("step = {}".format(i))
plt.savefig("mould_effect_without_angle_constraint.png", dpi=300)
plt.show()
```
